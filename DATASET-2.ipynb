```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/khuspreetkaur/Downloads/netflix.csv")
```

```python
#displaying first five rows of the data
df.head()
```

```python
#to check if there is any duplicate data
df[df.duplicated()]
```

```python
#count of movies and tv shows from each country
df['country'].value_counts()
```

```python
#count of movies and tv shows according to rating
df['rating'].value_counts()
```

```python
#plot showing rating with count
plt.figure(figsize=(12,8))
ax = sns.countplot(y="rating",data=df, order=df.rating.value_counts().index[0:15])
```

```python
#count of movies and tv shows according to duration
df['duration'].value_counts()
```

```python
#plot showing duration with count
plt.figure(figsize=(12,8))
ax = sns.countplot(y="duration",data=df, order=df.duration.value_counts().index[0:15])
```

```python
#transforming date_added column to yyyy-mm-dd format and extracting added year
date_yymmdd = pd.to_datetime(df['date_added'])
df['year_added'] = date_yymmdd.dt.year
```

```python
#showing the added data set with added column
df.head()
```

```python
#dataset with type just as tv shows
df_shows = df[df['type'] == "TV Show"]
df_shows
```

```python
#dataset with type just as movie
df_movies = df[df['type'] == "Movie"]
df_movies
```

```python
#movie and tv shows comparison through bar plot
fig = plt.figure(figsize=(8,6))
sns.countplot(x="type", data=df)
plt.show()
```

```python
#bar plot showing count of movies and tv shows through the years
df['year_added'].value_counts().plot(kind='bar')
```

```python
#plot showing the movies and tv shows according to the genres
plt.figure(figsize=(12,8))
ax = sns.countplot(y="listed_in",data=df, order=df.listed_in.value_counts().index[0:15])
```

```python
#sorted dataset in ascending order of year_added
df.sort_values(by=['year_added'])
```

```python
#movie and tv show count comparison for the year 2021
y21 = df[df['year_added'] == 2021.0]
fig = plt.figure(figsize=(8,6))
sns.countplot(x="type", data=y21)
plt.show()
```
